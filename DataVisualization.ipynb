```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
iris_df['species'] = iris_df['species'].apply(lambda x: iris.target_names[x])
```

```python
colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
species_colors = iris_df['species'].map(colors)
```

```python
plt.figure(figsize=(10, 6))
for species in iris_df['species'].unique():
    subset = iris_df[iris_df['species'] == species]
    plt.scatter(subset['petal length (cm)'], subset['petal width (cm)'], label=species, color=colors[species])
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Distribution of Iris Features')

# Sepal length
for species in iris_df['species'].unique():
    subset = iris_df[iris_df['species'] == species]
    axes[0, 0].hist(subset['sepal length (cm)'], alpha=0.5, label=species, color=colors[species], bins=20, density=True)
axes[0, 0].set_title('Sepal Length Distribution')
axes[0, 0].legend(title='Species')

# Sepal width
for species in iris_df['species'].unique():
    subset = iris_df[iris_df['species'] == species]
    axes[0, 1].hist(subset['sepal width (cm)'], alpha=0.5, label=species, color=colors[species], bins=20, density=True)
axes[0, 1].set_title('Sepal Width Distribution')
axes[0, 1].legend(title='Species')

# Petal length
for species in iris_df['species'].unique():
    subset = iris_df[iris_df['species'] == species]
    axes[1, 0].hist(subset['petal length (cm)'], alpha=0.5, label=species, color=colors[species], bins=20, density=True)
axes[1, 0].set_title('Petal Length Distribution')
axes[1, 0].legend(title='Species')

# Petal width

for species in iris_df['species'].unique():
    subset = iris_df[iris_df['species'] == species]
    axes[1, 1].hist(subset['petal width (cm)'], alpha=0.5, label=species, color=colors[species], bins=20, density=True)
axes[1, 1].set_title('Petal Width Distribution')
axes[1, 1].legend(title='Species')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
